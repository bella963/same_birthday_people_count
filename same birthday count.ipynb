## Simulation of how many people asked before getting a same birthday

Multiple experiments are hold. At each experiment, we asked as many people as possible until we get a match.  First generating a random value from 0 and 1 uniformly. The probability for a person's birthday is $1/365$. If next person asked having a different birthday, the probability is $2/365$.

```python


def same_birthday_count(num_exp):
    
    record_person = []
    
    exp = 0 #the number of experiment
    
    for i in range(num_exp):
        num_person = 1
        random_prob = np.random.uniform(0,1)
        while random_prob > num_person/365:
            num_person += 1
            random_prob = np.random.uniform(0,1)
            
        print(f"the number of person asked: {num_person}")
        
        record_person.append(num_person)
        exp += 1
        
        print(f"the nuber of experiment performed :{exp}")
        #print(f"the frequency of num_person/num_exp: {num_person/exp}")
    return np.mean(record_person)
    
```

```python
out = same_birthday_count(1000)
```
